```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("Titanic_data.csv")
```

```python
titanic_df.head()
```

## 1. Bar Graph

```python
titanic_df.Survived.value_counts().plot(
    kind = 'barh', figsize = (4,2), legend = True,
    title = "Passenger Survival Rate", grid = True,
    colormap = "Set1"
)
```

```python
import matplotlib.pyplot as plt

x = titanic_df.Survived.value_counts()
plt.bar(['Dead','Survived'], x.to_numpy())
plt.show()
plt.close()
```

### 2. Pie Chart


```python
titanic_df.Embarked.value_counts().plot.pie(
    legend = True, title= " Pie Chart", figsize = (4,4)
)
```

### 3. Line Graph


```python
titanic_df.Embarked.value_counts().plot(
    kind = "line",
    legend = True, title= " Line Graph", figsize = (4,4)
)
```

## 4. Scatter Graph (x,y)

```python
titanic_df.plot.scatter("Age","Fare")
```

### 5. Histogram

```python
titanic_df.Age.plot.hist(bins = 20)
```

```python
titanic_df.Age.plot.kde()
```

### 6. Heatmap

```python
import seaborn as sns
```

```python
correlation = titanic_df[["Pclass","Age","Fare","SibSp"]].corr()

sns.heatmap(correlation, annot=True, cmap="Spectral")
```
