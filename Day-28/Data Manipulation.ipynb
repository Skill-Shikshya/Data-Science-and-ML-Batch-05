## DataFrame Manupulation

### Inserting and Deleting Columns or Rows

```python
# Dropping a column
# axis : {0 or 'index', 1 or 'columns'}, default 0
#     Whether to drop labels from the index (0 or 'index') or columns (1 or 'columns')

titanic_df.drop("PassengerId", axis = 1, inplace=True)
```

```python
titanic_df.head()
```

```python
titanic_df.shape
```

```python
# Dropping rows
return_df = titanic_df.drop([0,2,4,6], axis = 0, inplace=False)
return_df.head()
```

```python
# Adding/ Creating a Column
last_names = [name.split(",")[0] for name in titanic_df.Name]
len(last_names)
```

```python
titanic_df["LastName"] = last_names
```

```python
titanic_df.head()
```

```python
salutation = [name.split(",")[1].split(".")[0].strip() for name in titanic_df.Name]
len(salutation)
```

```python
titanic_df["Salutation"] = salutation
```

```python
titanic_df.head()
```

```python
titanic_df.drop(["LastName","Salutation"], axis = 1, inplace=True)
```

```python
# Adding/ Creating a Rows
titanic_df.loc[0,"Cabin"] = "C23"
```

```python
titanic_df.head()
```

```python
titanic_df.loc[900,:] = [0,3,"Thapa, Mr. Ram","male", 33.7,0,1,"A100",33.4,"C34", "C"]
```

```python
titanic_df.loc[900,:]
```

```python
titanic_df.tail()
```

```python
titanic_df.drop(900,axis=0, inplace = True) 
```

### Handling a NaN or Null Value

```python
titanic_df.isna().sum()
```

```python
# Age : Imputation with Mean
age_mean = titanic_df.Age.mean()
titanic_df.Age.fillna(age_mean, inplace=True)
```
